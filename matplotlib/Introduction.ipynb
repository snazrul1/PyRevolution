Import libraries

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

$x = [0, 1, 2, 3, 4, 5, 6, ..... 100]$

$y1 = x^2$

$y2 = x^3$

```python
x = [i for i in range(101)]
y1 = [i**2 for i in x]
y2 = [i**3 for i in x]
```

PLOTTING!

```python
plt.figure(figsize=(15,5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.title('I Love Cats!', fontsize=16)
plt.xlabel('X AXIS', fontsize=13)
plt.ylabel('Y AXIS', fontsize=13)
plt.legend(['Quadratic', 'Cubic'], loc=0)
```
